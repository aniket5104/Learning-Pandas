# **groupby():**
- groupby() splits the data into groups based on some criteria, then lets you apply aggregate functions (like sum, mean, count, etc.) on each group.

```python
movies.head(1)
```

```python
genres = movies.groupby('Genre') #returns a DataFrameGroupBy object
```

## Applying **aggregate functions** on groupby objects

### **Common Aggregations**:
| Function         | Use                         |
| ---------------- | --------------------------- |
| `sum()`          | Sum of values               |
| `mean()`         | Average                     |
| `count()`        | Count of non-NaNs           |
| `min()`, `max()` | Min / Max                   |
| `std()`          | Standard deviation          |
| `agg()`          | Apply multiple aggregations |


```python
genres.sum(numeric_only=True) #sums up all numeric columns for each genre
```

```python
genres.min()
```

Find the top 3 genres by total earning.

```python
genres['Gross'].sum().sort_values(ascending=False).head(3)
```

Find the genre with highest imdb rating.

```python
genres['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

Find director with most popularity on the basis of votes.

```python
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

Find the highest rating of the highest rated movie of each genre

```python
genres['IMDB_Rating'].max()
```

Find no. of movies by each actor.

```python
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

## .groupby() **attributes** and **methods**:
1. len(): Find total no. of groups formed

```python
len(movies.groupby('Genre'))
```

---

2. .size(): Find no. of rows in each group

```python
movies.groupby('Genre').size()
```

---

3. first(), last() and nth(): 

```python
genres = movies.groupby('Genre')
```

```python
genres.first()
```

```python
genres.last()
```

```python
genres.nth(3)
```

---

4. get_group(): Get a specific group from the grouped object

```python
genres.get_group('Horror')
```

### **NOTE**: **.groups** returns a **dictionary** with the group names as keys and the indices of the rows in each group as values.

```python
genres.groups
```

---

5. **.describe()** : Returns a DataFrame with descriptive statistics for each group.

```python
genres.describe()
```

---

6. **.sample()** : Returns a random sample of rows from each group.

```python
genres.sample()
```

---

7. **nunique()**: Returns the number of unique values in each group.
