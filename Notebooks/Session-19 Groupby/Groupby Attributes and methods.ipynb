# **groupby():**
- groupby() splits the data into groups based on some criteria, then lets you apply aggregate functions (like sum, mean, count, etc.) on each group.

```python
movies.head(1)
```

```python
genres = movies.groupby('Genre') #returns a DataFrameGroupBy object
```

## Applying **aggregate functions** on groupby objects

### **Common Aggregations**:
| Function         | Use                         |
| ---------------- | --------------------------- |
| `sum()`          | Sum of values               |
| `mean()`         | Average                     |
| `count()`        | Count of non-NaNs           |
| `min()`, `max()` | Min / Max                   |
| `std()`          | Standard deviation          |
| `agg()`          | Apply multiple aggregations |


```python
genres.sum(numeric_only=True) #sums up all numeric columns for each genre
```

## **Groupby Aggregation Methods(agg)**

1. By passing a dictionary to the **.agg()** method, you can specify different aggregation functions for different columns.

```python
genres.agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating' : 'mean',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'min'
    }
)
```

2. by passing a list of functions to the **.agg()** method, you can apply multiple aggregation functions to the grouped data.

```python
genres.agg({
    'Runtime': ['min', 'max', 'mean'],
    'IMDB_Rating': ['min', 'max', 'mean'],
    'Gross': ['min', 'max', 'mean']
})
```

## **Looping on a groupby object:**

```python
for genre, group_df in genres:
    print("Group:", genre)
    print(group_df[['Series_Title', 'IMDB_Rating']])  # Show fewer columns
    print("\n")
```

Find the most rrated movie in each genre.

```python
dfs= []
for group, data in genres:
    dfs.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
df = pd.concat(dfs, ignore_index=True)
df
```

### **Split–Apply–Combine Paradigm (Pandas groupby)**

Q. Find number of movies starting with 'A' in each genre.

```python
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

---

Q. Find the ranking of each movie in the group according to IMDB score.

```python
def cal_rank(group):
    group['Rank'] = group['IMDB_Rating'].rank(ascending=False, method='min')
    return group
```

```python
genres.apply(cal_rank)
```

Q. Find the normalized IMDB rating for each movie in the group group wise.

```python
def nomal_rating(group):
    group['Normal_rating']=(group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(nomal_rating)
```
