# Boolean Indexing 
- Boolean indexing, also known as boolean masking, is a powerful feature in Pandas used to select subsets of data from a Series or DataFrame based on their actual values, rather than their labels or positions.

- It works by creating a "mask" â€“ a Series of True and False values. The elements corresponding to True are selected, and those corresponding to False are discarded.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/anike/OneDrive/Desktop/Learning Pandas/Datasets/Session-16/kohli_ipl.csv")
vk= df['runs']
```

```python
vk[vk > 50].size # Count of runs greater than 50
```

```python
vk[vk == 0].size # Count of runs equal to 0
```
